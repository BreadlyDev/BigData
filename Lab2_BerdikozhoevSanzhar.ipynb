```python
# 1.1 Создайте одномерный массив my_array, в котором по порядку все числа до 100, кратные 5
import numpy as np 

my_array = np.arange(5, 101, 5)
print(my_array)
```

```python
# 2.1 Из полученного массива получить массив с кратными 10 числами
array10 = np.extract(my_array % 10 == 0, my_array)
print(array10)
```

```python
# 3.1 Получить двумерный массив с двумя столбцами
array2d = array10.reshape(5, 2)
print(array2d)
```

```python
# 4.1 Получить двумерный массив с двумя столбцами
my_array2 = array2d - 5
print(my_array2)
```

```python
# 5.1 Соедините массивы, полученные в п.3.1 и 4.1, горизонтально. Используйте .T для транспонирования матрицы.
stacked_arrays = np.hstack([array2d, my_array2])
print(stacked_arrays)
```

```python
# 6.1 Разделите 2 и 3 строку на 4 верхних строки my_array. Используйте .T для транспонирования матрицы
div_res = stacked_arrays[1] / stacked_arrays [3] 
div_res2 = stacked_arrays[2] / stacked_arrays [3]

print(div_res)
print(div_res2)
```

```python
# 7.1. Оставить только те значения, которые не изменились после деления
stacked_arrays2 = stacked_arrays.copy().astype(float)
stacked_arrays2[1] = div_res
stacked_arrays2[2] = div_res2

print(f'{stacked_arrays} \n')

equal_rows = np.all(stacked_arrays == stacked_arrays2, axis=1)
unchanged = stacked_arrays2[equal_rows]

print(unchanged)
```

```python
# 8.1. Снова получить трехстрочный двумерный массив и поменять местами первую и третью строки
unchanged[[0, 2]] = unchanged[[2, 0]]
print(unchanged)
```

```python
# 9.1. Получить среднее, медиану и стандартное отклонение для каждой строки, используя команды np.mean(), np.median(), np.std().
mean_values = np.mean(unchanged, axis=1)
median_values = np.median(unchanged, axis=1)
std_values = np.std(unchanged, axis=1)

print('Mean: ', mean_values)
print('Median: ', median_values)
print('Std: ', std_values)
```

```python
# 10.1.	Сделать нормализацию на максимальное значение вдоль строк, поделив каждый элемент массива на max() вдоль строк.
maximum = np.max(unchanged, axis=1, keepdims=True)

norm_res = unchanged / maximum
print(f'{maximum} \n')
print(norm_res)
```

```python
# 2.6 Удалите строки, не содержащие значения в столбце «Total Population» - “-” 

df1=df[df['Total Population']!='-']
df1
```

```python
# 2.7 Получите описательную статистику по переменным (например, max(), min(), describe())

print(df1.describe())
print(df1.max())
print(df1.min())
print(df1['Total Population'].describe())
```

```python
# 2.8 Выполните сортировку данных по нескольким критериям

df2 = df1.sort_values(by=["Birth Rate", "Total Population", "Life Expectancy"])
print(df2)
```

```python
# Форматирование "Общей численности населения" для поиска

df1Fl = df1.copy()
df1Fl["Total Population"] = df1Fl["Total Population"].str.replace(',', '').astype(float)
df1Fl
```

```python
# 2.9 Выведите название страны, которая имеет максимальную численность населения
sorted_population = df1Fl.sort_values(by=["Total Population"])
print("Страна с максимальной численностью населения =", sorted_population["Country"].iloc[-1])
```

```python
# 2.10 	Выведите название страны, которая имеет минимальную численность населения
print("Страна, с минимальной численностью населения =", sorted_population["Country"].iloc[0])
```
